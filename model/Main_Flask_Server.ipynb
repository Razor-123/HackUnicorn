```python
model_name = "google/pegasus-cnn_dailymail"
device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer2 = AutoTokenizer.from_pretrained(model_name)
model2 = PegasusForConditionalGeneration.from_pretrained(model_name).to(device)
```

```python
from transformers import AutoTokenizer, XLMProphetNetForConditionalGeneration, XLMProphetNetConfig

config = XLMProphetNetConfig(ngram=2)
model=XLMProphetNetForConditionalGeneration(config)
model = XLMProphetNetForConditionalGeneration.from_pretrained('microsoft/xprophetnet-large-wiki100-cased-xglue-ntg')
tokenizer = AutoTokenizer.from_pretrained('microsoft/xprophetnet-large-wiki100-cased-xglue-ntg')
```

```python
from flask import Flask,request, jsonify,make_response
from flask_ngrok import run_with_ngrok
import speech_recognition as sr

app = Flask(__name__)
run_with_ngrok(app)   

  
@app.route("/",methods = ['GET'])
def home3():
  return "hello world!"

@app.route("/upload",methods = ['POST'])
def home():
  f = request.files['file']
  f.save(f.filename)
  
  r=sr.Recognizer()
  EN_SENTENCE = r.listen(f.filename)
  inputs = tokenizer([EN_SENTENCE], padding=True, max_length=256, return_tensors='pt')
  summary_ids = model.generate(inputs['input_ids'], num_beams=4, max_length=100, early_stopping=True)
  heading = tokenizer.batch_decode(summary_ids, skip_special_tokens=True)
  src_text = [EN_SENTENCE]
  batch = tokenizer2(src_text, truncation=True, padding="longest", return_tensors="pt").to(device)
  translated = model2.generate(**batch)
  notes=tokenizer2.batch_decode(translated, skip_special_tokens=True)
  tgt_text = tokenizer2.batch_decode(translated, skip_special_tokens=True)

  result = {
      "text":EN_SENTENCE,
      "heading":heading[0],
      "overview":tgt_text[0],
      "notes":notes,
  }
  return jsonify(result)

@app.route("/getmcqs",methods = ['GET'])
def home2():
  EN_Transcript = request.get_json()['transcript']
  res = []
  for i in range(3):
    question_and_answers = model.generate(EN_Transcript, num_beams=4, max_length=100, early_stopping=True)
    # Extract the question and answers from the generated model output
    lines = question_and_answers.strip().split(";")
    question = lines[0]
    answers = lines[1:]
    correct_answer=lines[:1]
    context = {
              'question': question,
              'answer1': answers[0],
              'answer2': answers[1],
              'answer3': answers[2],
              'answer4': answers[3],
              'correct_answer': correct_answer,
          }
    res.append(context)
  final_mcqs = jsonify(res)
  return make_response(final_mcqs)
  

    
if __name__ == "__main__":
    app.run()
```
